```python
import tensorflow as tf

# 1. Create tensors
scalar = tf.constant(10)
vector = tf.constant([1, 2, 3, 4, 5])
matrix = tf.constant([[1, 2], [3, 4]])

print("Scalar:", scalar.numpy(), "Shape:", scalar.shape)
print("Vector:", vector.numpy(), "Shape:", vector.shape)
print("Matrix:\n", matrix.numpy(), "Shape:", matrix.shape)

# 2. Tensor Operations
tensor_a = tf.constant([1, 2, 3])
tensor_b = tf.constant([4, 5, 6])

add_result = tf.add(tensor_a, tensor_b)
sub_result = tf.subtract(tensor_a, tensor_b)
mul_result = tf.multiply(tensor_a, tensor_b)

print("\nAddition:", add_result.numpy())
print("Subtraction:", sub_result.numpy())
print("Multiplication:", mul_result.numpy())

# 3. Reshape and Type Conversion
tensor_c = tf.constant([1, 2, 3, 4, 5, 6], dtype=tf.int32)
reshaped = tf.reshape(tensor_c, [2, 3])
converted = tf.cast(tensor_c, dtype=tf.float32)

print("\nReshaped (2x3):\n", reshaped.numpy())
print("Converted to float:", converted.numpy())

# 4. Statistical Operations
tensor_d = tf.constant([10, 20, 30, 40, 50], dtype=tf.float32)

sum_val = tf.reduce_sum(tensor_d)
mean_val = tf.reduce_mean(tensor_d)
max_val = tf.reduce_max(tensor_d)
min_val = tf.reduce_min(tensor_d)

print("\nSum:", sum_val.numpy())
print("Mean:", mean_val.numpy())
print("Max:", max_val.numpy())
print("Min:", min_val.numpy())
```

34. Train Model
Input: Same
Task: Train using fit() for multiple epochs.
Output: Loss decreases over epochs.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Input and output data as NumPy arrays
X = np.array([[1],[2],[3],[4]], dtype=float)
y = np.array([2,4,6,8], dtype=float)

# Build Sequential model with hidden (ReLU) and output layer
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(1,)),  # hidden layer
    layers.Dense(1)  # output layer
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model for multiple epochs
history = model.fit(X, y, epochs=50, verbose=0)

# Print first 10 loss values to show decrease
print("First 10 losses:", history.history['loss'][:10])
print("Final loss:", history.history['loss'][-1])
```

37. Binary Classification
Input: X=[[0],[1],[2],[3]], y=[0,0,1,1]
Task: Use sigmoid and binary_crossentropy.
Output: Predictions close to [0,0,1,1].

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Input data
X = np.array([[0],[1],[2],[3]], dtype=float)
y = np.array([0,0,1,1], dtype=float)

# Build binary classification model
model = keras.Sequential([
    layers.Dense(4, activation='relu', input_shape=(1,)),  # hidden layer
    layers.Dense(1, activation='sigmoid')  # output layer for binary classification
])

# Compile with binary crossentropy
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
model.fit(X, y, epochs=100, verbose=0)

# Predictions
predictions = model.predict(X)
print("Predictions:", predictions)
```

38. Multi-Class Classification
Input: X=[[1],[2],[3]], y=[0,1,2]
Task: Use softmax and sparse categorical crossentropy.
Output: Probability distribution.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Input data
X = np.array([[1],[2],[3]], dtype=float)
y = np.array([0,1,2], dtype=int)  # class labels

# Build multi-class classification model
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(1,)),  # hidden layer
    layers.Dense(3, activation='softmax')  # output layer for 3 classes
])

# Compile with sparse categorical crossentropy
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(X, y, epochs=100, verbose=0)

# Predictions
predictions = model.predict(X)
print("Predictions (probability distributions):")
print(predictions)
```

39. EarlyStopping
Input: Any dataset
Task: Use EarlyStopping callback.
Output: Training stops early.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping

# Example dataset
X = np.array([[1],[2],[3],[4]], dtype=float)
y = np.array([2,4,6,8], dtype=float)

# Build model
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(1,)),
    layers.Dense(1)
])

# Compile
model.compile(optimizer='adam', loss='mse')

# Define EarlyStopping callback
early_stop = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)

# Train with EarlyStopping
history = model.fit(X, y, epochs=200, callbacks=[early_stop], verbose=0)

print("Training stopped after", len(history.history['loss']), "epochs")
print("Final loss:", history.history['loss'][-1])
```

40. Save and Load Model
Input: Trained model
Task: Save and reload model.
Output: Same predictions after loading.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Training data
X = np.array([[1],[2],[3],[4]], dtype=float)
y = np.array([2,4,6,8], dtype=float)

# Build model
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(1,)),
    layers.Dense(1)
])

# Compile
model.compile(optimizer='adam', loss='mse')

# Train
model.fit(X, y, epochs=50, verbose=0)

# Save using the new Keras format
model.save("my_model.keras")

# Load the model back
loaded_model = keras.models.load_model("my_model.keras")

# Verify predictions
print("Original model prediction:", model.predict(np.array([[7]], dtype=float)))
print("Loaded model prediction:", loaded_model.predict(np.array([[7]], dtype=float)))
```
